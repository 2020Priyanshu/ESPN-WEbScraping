```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
import os
```

```python
url="https://www.espncricinfo.com/cricketers/team/india-6"
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
```

```python
cricketers_webpage = requests.get(url,headers).text
```

```python
cricketers = BeautifulSoup(cricketers_webpage,'lxml')
```

```python
cricketer_cards=[]
```

```python
for card in cricketers.find_all('div',class_='ds-flex ds-p-4 ds-flex-row ds-space-x-4 ds-border-line ds-border-b odd:ds-border-r last:ds-border-none'):
    cricketer_cards.append(card)
```

```python
# get names of all cricketers
cricketer_names=[]
cricketer_links=[]
for card in cricketer_cards:
    name=card.find('span',class_='ds-text-tight-l')
    link=card.find('a',attrs={'href':re.compile('/cricketers/')})
    cricketer_names.append(name.text)
    cricketer_links.append(link.get('href'))

cricketer_names,cricketer_links
```

```python
len(cricketer_names) == len(cricketer_links)
```

```python
base_url='https://www.espncricinfo.com'
for link in cricketer_links:
    cricketer_url =f"{base_url}{link}"
    print(cricketer_url)
```

```python
base_url='https://www.espncricinfo.com'
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
names = []
ages=[]
batting_styles=[]
bowling_styles=[]
playing_roles=[]
for link in cricketer_links:
    cricketer_url =f"{base_url}{link}"
    cricketer_webpage=requests.get(url=cricketer_url,headers=headers).text
    cricketer = BeautifulSoup(cricketer_webpage,'lxml')
    info_card = cricketer.find('div',class_='ds-grid lg:ds-grid-cols-3 ds-grid-cols-2 ds-gap-4 ds-mb-8')
    info=info_card.find_all('span',class_='ds-text-title-s ds-font-bold ds-text-typo')
    name = info[0].text
    age=info[2].text

    if len(info) == 5:
        batting_style=info[3].text
        bowling_style='NA'
        playing_role=info[4].text
    elif len(info) == 7:
        batting_style=info[4].text
        bowling_style=info[5].text
        playing_role=info[6].text
    elif len(info) == 8:
        batting_style=info[4].text
        bowling_style=info[5].text
        playing_role=info[7].text
    else:
        batting_style=info[3].text
        bowling_style=info[4].text
        playing_role=info[5].text

    if bowling_style == 'Wicketkeeper':
        bowling_style='NA'

    names.append(name)
    ages.append(age)
    batting_styles.append(batting_style)
    bowling_styles.append(bowling_style)
    playing_roles.append(playing_role)

```

```python
cricketers_data=pd.DataFrame({'name':names,'age':ages,'batting_style':batting_styles,'bowling_style':bowling_styles,'playing_roles':playing_roles})
```
